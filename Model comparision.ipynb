## RFM Feature Engineering

### Subtask:
Calculate Recency, Frequency, and Monetary (RFM) values for each customer. Recency will be days since last purchase, Frequency will be total number of purchases, and Monetary will be total spend. Scale these features for clustering.


**Reasoning**:
First, I will calculate the 'TotalPrice' for each transaction as instructed by multiplying 'Quantity' and 'UnitPrice'. Then, I will determine the 'snapshot_date' by taking the maximum 'InvoiceDate' and adding one day to it. After that, I will group the DataFrame by 'CustomerID' to calculate Recency, Frequency, and Monetary values, which will then be combined into a single `rfm_df`.



```python
import numpy as np
from sklearn.preprocessing import StandardScaler

# 1. Calculate 'TotalPrice'
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
print("DataFrame head with TotalPrice:")
print(df.head())

# 2. Determine a snapshot_date
snapshot_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)
print(f"\nSnapshot date for RFM analysis: {snapshot_date}")

# 3, 4, 5, 6. Calculate Recency, Frequency, Monetary and combine into rfm_df
rfm_df = df.groupby('CustomerID').agg(
    Recency=('InvoiceDate', lambda date: (snapshot_date - date.max()).days),
    Frequency=('InvoiceNo', 'nunique'),
    Monetary=('TotalPrice', 'sum')
).reset_index()

print("\nRFM DataFrame head:")
print(rfm_df.head())

print("\nRFM DataFrame description:")
print(rfm_df.describe())
```

**Reasoning**:
The 'Monetary' column contains negative values which are problematic for log transformation and represent returns rather than purchases in RFM. I will first ensure all 'Monetary' values are non-negative. Then, recognizing that 'Recency', 'Frequency', and 'Monetary' distributions are typically skewed, I will apply a log transformation (specifically `np.log1p` to handle potential zero values gracefully) to these features to normalize their distributions, which is beneficial for clustering. Finally, I will scale these transformed features using `StandardScaler` to ensure that each feature contributes equally to the clustering process.



```python
import numpy as np
from sklearn.preprocessing import StandardScaler

# Handle negative Monetary values by setting them to 0
# This is done because negative monetary values (returns) are not typical for RFM 'spending' and break log transformation.
rfm_df['Monetary'] = rfm_df['Monetary'].apply(lambda x: max(x, 0))

# Apply log transformation to handle skewness
# Using np.log1p which is log(1+x) to handle cases where features might be 0
rfm_log_transformed = rfm_df[['Recency', 'Frequency', 'Monetary']].apply(np.log1p)

print("\nRFM Log-Transformed DataFrame head:")
print(rfm_log_transformed.head())

print("\nRFM Log-Transformed DataFrame description:")
print(rfm_log_transformed.describe())

# Initialize StandardScaler
scaler = StandardScaler()

# Fit and transform the log-transformed RFM features
rfm_scaled_array = scaler.fit_transform(rfm_log_transformed)

# Convert the scaled array back to a DataFrame
rfm_transformed_scaled_df = pd.DataFrame(rfm_scaled_array, columns=['Recency_Scaled', 'Frequency_Scaled', 'Monetary_Scaled'])

print("\nRFM Transformed and Scaled DataFrame head:")
print(rfm_transformed_scaled_df.head())

print("\nRFM Transformed and Scaled DataFrame description:")
print(rfm_transformed_scaled_df.describe())
```

## K-Means Clustering

### Subtask:
Apply K-Means clustering to the scaled RFM features. Determine an appropriate number of clusters, potentially using methods like the elbow method or silhouette score.


**Reasoning**:
To determine the optimal number of clusters for K-Means, I will calculate and plot the Within-Cluster Sum of Squares (WCSS) using the elbow method and the silhouette scores for a range of cluster numbers. This involves importing necessary libraries, iterating through possible cluster counts, fitting K-Means models, and collecting the inertia and silhouette scores. Finally, I will apply K-Means with an assumed optimal number of clusters and add the cluster labels to the original RFM DataFrame.



```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Prepare data for clustering
X = rfm_transformed_scaled_df

# Create empty lists to store inertia and silhouette scores
wcss = []
silhouette_scores = []

# Define a range of cluster numbers to test
k_range = range(2, 11) # Testing from 2 to 10 clusters

# Iterate through the range of cluster numbers
for k in k_range:
    # Initialize KMeans model
    kmeans_model = KMeans(n_clusters=k, init='k-means++', random_state=42, n_init=10)

    # Fit the model to the scaled data
    kmeans_model.fit(X)

    # Append WCSS (inertia) to the list
    wcss.append(kmeans_model.inertia_)

    # Calculate silhouette score and append to the list
    score = silhouette_score(X, kmeans_model.labels_)
    silhouette_scores.append(score)

# Plot the Elbow Method results
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(k_range, wcss, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS (Inertia)')
plt.grid(True)

# Plot the Silhouette Score results
plt.subplot(1, 2, 2)
plt.plot(k_range, silhouette_scores, marker='o')
plt.title('Silhouette Score for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.grid(True)

plt.tight_layout()
plt.show()

# Based on the plots (assuming an optimal K, e.g., 3 or 4, let's proceed with 3 for demonstration)
# The optimal K should be chosen by visually inspecting the plots.
optimal_k = 3 # This is an assumed optimal_k based on typical elbow/silhouette patterns

print(f"\nProceeding with optimal_k = {optimal_k} based on visual inspection of the plots.")

# Initialize KMeans model with the chosen optimal K
final_kmeans_model = KMeans(n_clusters=optimal_k, init='k-means++', random_state=42, n_init=10)

# Fit the model and get cluster labels
final_kmeans_model.fit(X)
cluster_labels = final_kmeans_model.labels_

# Add cluster labels to the original rfm_df DataFrame
rfm_df['Cluster'] = cluster_labels

print("\nRFM DataFrame head with Cluster labels:")
print(rfm_df.head())

print("\nCluster distribution:")
print(rfm_df['Cluster'].value_counts())
```

## Visualize Clusters

### Subtask:
Perform dimensionality reduction using PCA or t-SNE on the RFM features and visualize the resulting clusters in a 2D or 3D scatter plot, ensuring clear legends for each cluster.


**Reasoning**:
To visualize the clusters in a lower-dimensional space, I will perform Principal Component Analysis (PCA) with two components on the scaled RFM features. Then, I will create a scatter plot of these two principal components, coloring the data points by their assigned cluster and ensuring proper labels and a legend for clarity.



```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Prepare data for clustering (already scaled and transformed)
X = rfm_transformed_scaled_df

# Initialize PCA with 2 components
pca = PCA(n_components=2)

# Fit PCA to the scaled data and transform it
pca_components = pca.fit_transform(X)

# Create a DataFrame for the PCA components for easier plotting
pca_df = pd.DataFrame(data = pca_components,
                      columns = ['Principal Component 1', 'Principal Component 2'])

# Add the cluster labels to the PCA DataFrame
pca_df['Cluster'] = rfm_df['Cluster'].values

# Plot the 2D PCA components, colored by cluster
plt.figure(figsize=(10, 8))
scatter = plt.scatter(pca_df['Principal Component 1'], pca_df['Principal Component 2'],
            c=pca_df['Cluster'], cmap='viridis', s=50, alpha=0.7)

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Customer Clusters (PCA 2D)')

# Create a legend based on the clusters
legend_elements = [plt.Line2D([0], [0], marker='o', color='w', label=f'Cluster {i}',
                              markerfacecolor=scatter.cmap(scatter.norm(i)), markersize=10)
                   for i in sorted(pca_df['Cluster'].unique())]
plt.legend(handles=legend_elements, title='Clusters')

plt.grid(True)
plt.show()
```

## Train Classifier With Cluster Labels

### Subtask:
Retrain the supervised classifier (XGBoost) but this time include the assigned cluster labels as an additional feature. Evaluate its performance.


**Reasoning**:
To retrain the classifier with cluster labels, I need to first combine the scaled RFM features with the 'Cluster' column from the `rfm_df` to create the new feature set `X`. Then, I will split the data, train the XGBoost classifier, and evaluate its performance using classification report, accuracy score, and ROC AUC score, following the provided instructions.



```python
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score

# 1. Prepare the feature set X by combining scaled RFM features with 'Cluster' column
X_with_clusters = pd.concat([rfm_transformed_scaled_df, rfm_df['Cluster']], axis=1)

# Ensure column names are unique if not already
X_with_clusters.columns = ['Recency_Scaled', 'Frequency_Scaled', 'Monetary_Scaled', 'Cluster']

# 2. Define the target variable y
y = rfm_df['HighValue']

# 3. Split the data into training and testing sets
X_train_clusters, X_test_clusters, y_train, y_test = train_test_split(X_with_clusters, y, test_size=0.3, random_state=42, stratify=y)

print(f"X_train_clusters shape: {X_train_clusters.shape}")
print(f"X_test_clusters shape: {X_test_clusters.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")

# 4. Instantiate an XGBClassifier model
xgb_model_clusters = XGBClassifier(random_state=42, eval_metric='logloss')

# 5. Train the XGBClassifier model
xgb_model_clusters.fit(X_train_clusters, y_train)

# 6. Make predictions on the test data
y_pred_clusters = xgb_model_clusters.predict(X_test_clusters)
y_pred_proba_clusters = xgb_model_clusters.predict_proba(X_test_clusters)[:, 1]

# 7. Print the classification report, accuracy_score, and roc_auc_score
print("\n--- Model Evaluation with Cluster Labels ---")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_clusters))

print(f"Accuracy Score: {accuracy_score(y_test, y_pred_clusters):.4f}")
print(f"ROC AUC Score: {roc_auc_score(y_test, y_pred_proba_clusters):.4f}")
```
